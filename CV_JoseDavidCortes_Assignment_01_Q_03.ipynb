# Thresholding and Adaptative Thresholding

Lets import the required libraries and read the pictures. 
It is mandatory to read the image in grayscale format to perform thresholding.

```python
import cv2 
from matplotlib import pyplot as plt
imPath = '/content/drive/MyDrive/ColabNotebooks/image/rocket_grey.jpeg'
image = cv2.imread(imPath,0)
```

```python
# Values below 149 goes to 0 (black, everything above goes to 255 (white)
_,th0 = cv2.threshold(image,149,255,cv2.THRESH_BINARY) 
```

**Is there a better way off thresholding?**

The biggest downfall of those simple threshold methods is that we need to provide the threshold value (i.e. the 149 value we used previously).

**What if there was a smarter way of doing this?**

There is with, Adaptive thresholding.

```python
# Using adaptive Threshold MEAN
th1 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2) 

# Using adaptive Threshold GAUSSIAN
th2 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,15,5) 
```

Lets print the image after apply the threshold

```python
titles = ['Space X - Original Image','Simple Binary Thresholding','Adaptive Mean Thresholding','Adaptive Gaussian Thresholding']
images = [image,th0, th1, th2]
plt.figure(figsize=(20,10))
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```
